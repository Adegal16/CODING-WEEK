## Contribution : Exploratory Data Analysis (EDA)

Dans cette étape, j’ai réalisé une analyse exploratoire des données sur le dataset **Heart Failure Clinical Records** afin de comprendre la structure du jeu de données et d’identifier d’éventuels problèmes avant la phase de traitement.

J’ai commencé par charger le dataset et examiner ses caractéristiques générales, notamment le nombre d’observations, les types de variables et les statistiques descriptives. Cette première analyse a permis d’obtenir une vue d’ensemble des variables médicales présentes, comme l’âge, la fraction d’éjection, la créatinine sérique et le sodium sérique.

Ensuite, j’ai vérifié la présence de **valeurs manquantes** dans les données. L’analyse a montré que le dataset est **complet et ne contient pas de valeurs manquantes**, ce qui simplifie les étapes de prétraitement.

Une analyse des **valeurs aberrantes (outliers)** a également été réalisée sur les variables numériques à l’aide de visualisations de type **boxplot**. Des outliers ont été observés dans certaines variables comme **platelets**, **serum_creatinine** et **creatinine_phosphokinase**. Ces valeurs ont été conservées car elles peuvent correspondre à des cas médicaux réels.

J’ai ensuite étudié la **distribution de la variable cible (`DEATH_EVENT`)** afin d’évaluer l’équilibre des classes. L’analyse montre que le dataset est légèrement déséquilibré, avec environ **68 % de patients survivants contre 32 % de patients décédés**.

Une **analyse de corrélation** a été réalisée à l’aide d’une matrice de corrélation afin d’identifier les relations entre les variables. Cette analyse a mis en évidence certaines variables ayant un impact important sur la prédiction du décès, notamment **ejection_fraction**, **serum_creatinine** et **time**, qui présentent les corrélations les plus marquées avec la variable cible.

Enfin, j’ai analysé la distribution de certaines variables selon la classe de la variable cible afin de comparer les caractéristiques des patients survivants et décédés. Cette étape permet d’identifier les variables les plus discriminantes pour les étapes de modélisation.

🔬 Dataset

- **Source :** [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Heart+failure+clinical+records)
- **Taille :** 299 patients × 13 variables cliniques
- **Cible :** `DEATH_EVENT` (0 = survécu, 1 = décédé)
- **Variables :** âge, anémie, créatinine phosphokinase, diabète, fraction d'éjection, hypertension, plaquettes, créatinine sérique, sodium sérique, sexe, tabagisme, suivi temporel



## Contribution : Data Processing

Après l’analyse exploratoire des données (EDA), j’ai réalisé la phase de **prétraitement des données (data processing)** afin de préparer le dataset pour les étapes d’analyse et de modélisation.

La première étape a consisté à vérifier et confirmer l’absence de **valeurs manquantes** dans le dataset. Comme l’EDA l’a montré, les données sont complètes, ce qui a permis d’éviter des opérations d’imputation ou de suppression de lignes.

Ensuite, Une attention particulière a également été portée aux **valeurs aberrantes (outliers)** détectées lors de l’EDA, notamment dans les variables **platelets**, **serum_creatinine** et **creatinine_phosphokinase**. Ces valeurs ont été conservées car elles peuvent correspondre à des situations médicales réelles et donc contenir une information importante pour la modélisation.

Ce pipeline enchaîne cinq étapes structurées : le chargement du dataset depuis l'URL UCI, la détection des valeurs manquantes avec localisation précise (ligne et colonne), la détection des outliers par la méthode des ±2 écarts-types sur sept colonnes numériques, le split stratifié 80/20 avec export des fichiers `train.csv` et `test.csv`, et enfin l'équilibrage des classes par SMOTE appliqué exclusivement sur le jeu d'entraînement.

##  Ingénierie des Prompts — Tests Automatisés
L'une des problématiques rencontrées lors du développement du pipeline concernait le comportement du SMOTE sur les variables médicales à valeurs entières. En effet, après une première demande de génération du code d'équilibrage, le résultat produit conservait les valeurs interpolées telles quelles, générant des entrées comme `age = 67.18` ou `ejection_fraction = 38.7`, ce qui est médicalement incohérent pour des variables qui ne peuvent prendre que des valeurs entières. Ce problème était directement lié à la formulation initiale du prompt, qui ne précisait pas la contrainte d'intégrité sur ces colonnes.

Un second prompt, plus précis, a permis de corriger ce comportement en spécifiant explicitement que `age` doit rester une valeur entière et en demandant l'ajout d'un arrondi post-SMOTE suivi d'une conversion de type. Cette itération sur le prompt illustre l'importance de préciser les contraintes métier lors de la génération automatisée de code : une instruction vague produit un code fonctionnel mais sémantiquement incorrect, tandis qu'un prompt enrichi des règles du domaine médical aboutit à un résultat cohérent avec la réalité des données.
